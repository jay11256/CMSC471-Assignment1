```python
# Setup
import pandas as pd
df = pd.read_csv("./data/weather.csv")
```

```python
# Displaying data format/metadata
print(df.columns)
print(df.shape)
df.head()
```

```python
# Keeping US states only
print(df["state"].unique())

# US states
us_states = [
    'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA',
    'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD',
    'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ',
    'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC',
    'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY'
]

# Keep only rows where the state column is in US states
df = df[df['state'].isin(us_states)]
print(df["state"].nunique())
```

```python
# Looking into stations
stations_per_state = df.groupby('state')['station'].nunique()
print(stations_per_state)
```

```python
import pandas as pd

# Example: integer dates

# Convert integers to string first, then to datetime
# If your column is called 'date', first cast to string of just the numbers
df['date_int'] = df['date'].astype(str).str[-8:]  # last 8 digits: YYYYMMDD
df['date'] = pd.to_datetime(df['date_int'].astype(str), format='%y-%m-%d')

# Full year of 2017
all_dates = pd.date_range(start='2017-01-01', end='2017-12-31')

# Dates actually in the dataset
dataset_dates = df['date'].drop_duplicates()

# Missing dates
missing_dates = all_dates.difference(dataset_dates)

print("Number of missing days:", len(missing_dates))
missing_dates.to_series().to_csv('missing_dates_2017.csv', index=False)
```
